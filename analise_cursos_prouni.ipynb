#AGRUPANDO ITENS

```python

agrupando = df.groupby(['uf_busca'])
```

```python
agrupando['curso_busca'].value_counts()
```

```python
#agg agregador
df.groupby(['uf_busca', 'curso_busca', 'grau', 'turno']).agg(['mean', 'median'])
```
