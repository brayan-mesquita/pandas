```python
import pandas as pd
```

# LENDO O DATAFRAME EM CSV

```python
df = pd.read_csv('https://raw.githubusercontent.com/brayan-mesquita/Pandas/main/cursos-prouni.csv')
df.head()
```

```python
df[df['nome'] == 'Medicina'].describe()
```

#ORDENANDO O DATAFRAME

```python

df.sort_values(by=['mensalidade', 'bolsa_integral_ampla'], ascending=True).head(3)
```

```python
#CONTAGEM OS TIPOS DE CURSOS
df['nome'].value_counts()
```

#AGRUPANDO ITENS

```python

agrupando = df.groupby(['uf_busca'])
```

```python
agrupando['curso_busca'].value_counts()
```

```python
#agg agregador
df.groupby(['uf_busca', 'curso_busca', 'grau', 'turno']).agg(['mean', 'median'])
```

```python
#agrupar a UF e o curso buscado e contar a quantidade de bolsas
bolsas_por_estado = df.groupby(['uf_busca', 'curso_busca'])['bolsa_integral_cotas'].count()
```
