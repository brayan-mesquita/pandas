```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# data frame sobre tamanhos de arvores

```python
#df = pd.read_csv('/Users/brayan/Documents/cursos/FormacaoCD/13.PraÃÅtica em Python/dados/trees.csv')
df = pd.read_csv('https://raw.githubusercontent.com/brayan-mesquita/pandas/main/dados/trees.csv')
df.columns = ['cir', 'alt', 'vol']
df.head()
```

## plt

```python
plt.scatter(df.alt, df.vol, color='green', facecolors = 'none', marker='o')
plt.title('rela√ßao entre altitude e circunfer√™ncia')
plt.xlabel('Circunfer√™ncia')
plt.ylabel('Altitude')
```

```python
plt.scatter(df.vol, df.cir, color='blue', facecolors = 'none', marker='o')
plt.title('rela√ßao entre Volume e Circufer√™ncia')
plt.xlabel('Volume')
plt.ylabel('Circufer√™ncia')
```

```python
plt.plot(df.cir, df.vol)
plt.title('rela√ßao entre circunferencia e volume')
plt.xlabel('Circunferencia')
plt.ylabel('Volume')
```
